## Helpers 

```python
import boto3
import uuid
import json
from IPython.display import display, Markdown

AWS_REGION=boto3.session.Session().region_name

# Create Bedrock AgentCore client
client = boto3.client("bedrock-agentcore", region_name=AWS_REGION)

def invoke_agent_runtime(agent_arn, application_id, session_id, user_id, prompt, extra=None):
    jpayload = {"prompt": prompt, "user_id": user_id}
    if not(application_id is None):
        jpayload["application_id"] = application_id
    if not(extra is None):
        for k in extra:
            jpayload[k]=extra[k]
    return client.invoke_agent_runtime(
        agentRuntimeArn=agent_arn,
        runtimeSessionId=session_id,
        runtimeUserId=user_id,
        payload=json.dumps(jpayload),
        qualifier="DEFAULT" # Optional
    )


def print_response(response):
    streaming_body=response['response']
    raw_bytes = streaming_body.read()
    text_content = raw_bytes.decode('utf-8')
    j=json.loads(text_content)
    text_response=j.pop("response")
    print(json.dumps(j, indent=2))
    print("")
    print("")
    display(Markdown(text_response))
```

## Test Driver

```python
import random

def choose_item_from_list(l):
    idx = random.randint(0, len(l) - 1)
    return l[idx]

def run_one_invoke(arn, application_id, session_id, user_id, q, counts):
    try:
        print(q)
        resp=invoke_agent_runtime(arn, application_id, session_id, user_id, q)
        counts['good'] = counts['good'] + 1
        return True
    except Exception as e:
        print("Error")
        counts['bad'] = counts['bad'] + 1
        return False
        
```

## Trusteeship Tests

```python
arn = TRUSTEESHIP_AGENT_ARN
NUM_APPLICATIONS = 50
MAX_STAGE_ATTEMPTS = 5
PERSONAL_ISSUES = [
    "I had a criminal conviction for disturbing the peace ten years ago",
    "I served ten years for fraud and embezzlement",
    "I can only give two personal references but I think I need three",
    "I had some late payments a few years back which led to a low credit score",
    "I have a previous bankruptcy",
]
APPLICATION_REQUEST = "I am submitting my application for trusteeship"
APPLICATION_REQUEST_ISSUE = "I have a question. Will my application for trusteeship succeed given the following personal issue : {issue}"
MEDICAL_REQUEST = "Attached is the medical evaluation of the beneficiary"
CREDIT_REQUEST = "Credit check details provided"
CRIMINAL_REQUEST = "Criminal check details provided"
REFS_REQUEST = "Personal reference details provided"
STATUS_TYPE_TO_EV_TYPE = {
    "criminal_check": "CriminalCheckUpdate",
    "medical": "MedicalUpdate",
    "application": "ApplicationUpdate",
    "credit_check": "CreditCheckUpdate",
    "personal_references": "PersonalReferencesUpdate",
}
STATUS_TYPE_TO_Q = {
    "criminal_check": CRIMINAL_REQUEST,
    "medical": MEDICAL_REQUEST,
    "application": APPLICATION_REQUEST,
    "credit_check": CREDIT_REQUEST,
    "personal_references": REFS_REQUEST,
}
USERS = ["joecool", "snoopy", "redbaron"]
counts = {"good": 0, "bad": 0}

print(arn)
for i in range(NUM_APPLICATIONS):
    application_id = str(uuid.uuid4())
    session_id = str(
        uuid.uuid4()
    )  # will put all in a single session; not realistic, but PM doesn't care
    print("")
    print("")
    print(f"Application {application_id}")
    applicant_id = choose_item_from_list(USERS)
    user_id = "joecool"
    beneficiary_id = "auldlang"
    trustee_detail = {"applicant_id": applicant_id, "beneficiary_id": beneficiary_id}
    detail = trustee_detail
    status = {}
    app_go = True

    def run_trusteeship_invoke(q, detail):
        try:
            global status
            print(f"{q} {detail}")
            response = invoke_agent_runtime(
                arn, application_id, session_id, user_id, q, detail
            )
            streaming_body = response["response"]
            raw_bytes = streaming_body.read()
            text_content = raw_bytes.decode("utf-8")
            j = json.loads(text_content)
            text_response = j.pop("response")
            status = j
            print(status)
            return True
        except Exception as e:
            print("Error")
            counts["bad"] = counts["bad"] + 1
            return False

    def inject_until_good(q, detail, status_type, second_try_q, second_try_detail):
        stage_attempts = 0
        while not ("step_statuses" in status) or not (
            status["step_statuses"][status_type] == "complete"
        ):
            if stage_attempts > 0:
                q = second_try_q
                detail = second_try_detail
            print("")
            run_trusteeship_invoke(q, detail)
            stage_attempts += 1
            if stage_attempts > MAX_STAGE_ATTEMPTS:
                return False
        return True

    # application phase
    has_issue = random.randint(0, 5) == 5
    issue = choose_item_from_list(PERSONAL_ISSUES)
    q = (
        APPLICATION_REQUEST_ISSUE.format(issue=issue)
        if has_issue
        else APPLICATION_REQUEST
    )
    app_go = inject_until_good(
        q,
        detail,
        "application",
        f"{APPLICATION_REQUEST}. All required documents attached.",
        {"event": {"type": "ApplicationUpdate"}},
    )

    if app_go:
        # medical or criminal; the latter is generally not allowed at this stage
        early_criminal = random.randint(0, 20) == 20
        status_type = "criminal_check" if early_criminal else "medical"
        ev_type = STATUS_TYPE_TO_EV_TYPE[status_type]
        q = STATUS_TYPE_TO_Q[status_type]
        detail = {"event": {"type": ev_type}}
        app_go = inject_until_good(
            q, detail, status_type, f"{q}. All required documents attached.", detail
        )

    if app_go and early_criminal:
        # now time for medical
        status_type = "medical"
        ev_type = STATUS_TYPE_TO_EV_TYPE[status_type]
        q = STATUS_TYPE_TO_Q[status_type]
        detail = {"event": {"type": ev_type}}
        app_go = inject_until_good(
            q, detail, "medical", f"{q}. All required documents attached.", detail
        )

    if app_go:
        remaining = ["credit_check", "personal_references"] if early_criminal else ["criminal_check", "credit_check", "personal_references"]
        random.shuffle(remaining)
        for status_type in remaining:
            ev_type = STATUS_TYPE_TO_EV_TYPE[status_type]
            q = STATUS_TYPE_TO_Q[status_type]
            detail = {"event": {"type": ev_type}}
            app_go = inject_until_good(
                q,
                detail,
                status_type,
                f"{q}. All required documents attached.",
                detail,
            )
```
